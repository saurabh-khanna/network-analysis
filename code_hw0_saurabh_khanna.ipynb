```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1.4
h = g.to_undirected()
h.number_of_edges()
```

```python
#1.5
g.number_of_edges() - h.number_of_edges()
```

```python
#1.6
outdeg = g.out_degree()
outdeg_zero_list = [n for (n, deg) in outdeg if deg == 0]
len(outdeg_zero_list)
```

```python
#1.7
indeg = g.in_degree()
indeg_zero_list = [n for (n, deg) in indeg if deg == 0]
len(indeg_zero_list)
```

```python
#1.8
outdeg_list = [n for (n, deg) in outdeg if deg > 10]
len(outdeg_list)
```

```python
#1.9
indeg_list = [n for (n, deg) in indeg if deg < 10]
len(indeg_list)
```

```python
out_list = list(outdeg)
out_deg_list = [t[1] for t in out_list]
```

```python
pairs = []
for i in range(min(out_deg_list), max(out_deg_list) + 1):
    if i in out_deg_list and i != 0:
        tup = [i, out_deg_list.count(i)]
        pairs.append(tup)
```

```python
pairs = np.log10(np.array(pairs))
x = pairs[:,0]
y = pairs[:,1]
plt.plot(x, y)
plt.savefig("answer_2_1.png")
```

```python
coefs = np.polyfit(x,y,1)
ffit = np.poly1d(coefs)
coefs
```

```python
fig1 = plt.figure()                                                                                           
ax1 = fig1.add_subplot(111)                                                                                   
ax1.scatter(x, y)                                                                     
ax1.plot(x, ffit(x),color='black')                                                                     
plt.show()
```

```python
gr = nx.read_edgelist("stackoverflow-Java.txt", create_using=nx.DiGraph(), nodetype = int)
```
