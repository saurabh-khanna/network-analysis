```python
g = nx.Graph()
g.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 3), (2, 4), (3, 4)])
g.add_edges_from([(5, 6), (5, 7), (5, 8), (6, 7), (6, 8), (7, 8)])
g.add_edges_from([(9, 10), (9, 11), (9, 12), (10, 11), (10, 12), (11, 12)])
g.add_edges_from([(13, 14), (13, 15), (13, 16), (14, 15), (14, 16), (15, 16)])
g.add_edges_from([(4, 5), (13, 12), (1, 16), (8, 9)])
```

```python
nx.draw(g, with_labels = True)
```

```python
greedy_modularity_communities(g)
```

```python
partition = community.best_partition(g)
modularity = community.modularity(partition, g)
modularity
```

```python
values = [partition.get(node) for node in g.nodes()]
nx.draw_spring(g, cmap=plt.get_cmap('jet'), node_color = values, node_size=300, with_labels=True)
```
